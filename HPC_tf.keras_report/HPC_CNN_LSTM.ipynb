# Household Power Consumption(HPC) dataset-CNN-LSTM
&emsp;&emsp;本章主要回报对于家庭用电量数据集（HPC）的时间序列预测模型，在tf.keras框架下的多个模型复现训练。原始教学材料请见[How to Develop Convolutional Neural Networks for Multi-Step Time Series Forecasting](https://machinelearningmastery.com/how-to-develop-convolutional-neural-networks-for-multi-step-time-series-forecasting/)。  
&emsp;&emsp;HPC数据集包括以下7个时间序列变量，下载地址[household_power_consumption.zip](https://archive.ics.uci.edu/ml/machine-learning-databases/00235/household_power_consumption.zip)：
- **global_active_power**: The total active power consumed by the household (kilowatts).
- **global_reactive_power**: The total reactive power consumed by the household (kilowatts).
- **voltage**: Average voltage (volts).
- **global_intensity**: Average current intensity (amps).
- **sub_metering_1**: Active energy for kitchen (watt-hours of active energy).
- **sub_metering_2**: Active energy for laundry (watt-hours of active energy).
- **sub_metering_3**: Active energy for climate control systems (watt-hours of active energy).  
通过简单的减法计算，得到第8个变量，与`global_active_power`，`submetering_1`，`sub_metering_2`，`sub_metering_3`有线性关系如下：
```python
sub_metering_4 = (global_active_power * 1000 / 60) - (sub_metering_1 + sub_metering_2 + sub_metering_3)
```

## 数据预处理
&emsp;&emsp;数据预处理主要包括以下步骤：
- 缺失值填补（用前一日同一时刻的数据）
- 将分钟频率数据缩减为日频数据（日内加和）
- 将日频数据按照Sun.~Sat.一周分组
- 将数据切分为train, test数据集（2010年以前数据为train set, 2010年以后数据为test set）
- 将数据集整理为输入与输出，其中输入变量为当前8个变量两周共14天的数据，输出变量（标签）为**global_active_power**之后一周7天的数据
- 将train和test储存为**TFRecord**格式的文件，以便读取为**tf.Dataset**

````python
def write2tfrecord(lag=lag):
    # ------------------------------------------------
    # 将数据集切分为train/test部分，
    # 并且在train和test内部按week进行组合
    # ------------------------------------------------
    daily_data = pd.read_csv("household_power_consumption_days.csv",
                             header=0,
                             infer_datetime_format=True,
                             parse_dates=["datetime"],
                             index_col=["datetime"])

    train, test = split_dataset(daily_data.values)
    global n_train
    global n_test
    n_train = len(train) - lag
    n_test = len(test) - lag
    train_writer = tf.python_io.TFRecordWriter(train_file)
    test_writer = tf.python_io.TFRecordWriter(test_file)
    # train dataset tfrecord
    for i in range(len(train) - lag):
        vars_byt = []
        vars_byt.append(train[i + lag, :, 0].tostring())
        for k in range(8):
            vars_byt.append(train[i:(i + lag), :, k].reshape((-1,)).tostring())
        train_examples = tf.train.Example(features=tf.train.Features(
            feature=dict(zip(vars,
                             list(map(lambda x: tf.train.Feature(bytes_list=tf.train.BytesList(value=[x])),
                                      vars_byt))))

        ))
        train_writer.write(train_examples.SerializeToString())
    train_writer.close()

    # test dataset tfrecord
    for i in range(len(test) - lag):
        vars_byt = []
        vars_byt.append(test[i+lag, :, 0].tostring())
        for k in range(8):
            vars_byt.append(test[i:(i+lag), :, k].reshape((-1,)).tostring())
        test_examples = tf.train.Example(features=tf.train.Features(
            feature=dict(zip(vars,
                     list(map(lambda x: tf.train.Feature(bytes_list=tf.train.BytesList(value=[x])),
                              vars_byt))))

        ))
        test_writer.write(test_examples.SerializeToString())
    test_writer.close()
````

需要实现对**TFRecord**格式文件的解析函数

````python
# ---------------------------------------------
# 从TFRecord文件中读取并创建Dataset的parsing函数
# ---------------------------------------------
def parse_example(serial_example, var_indices=None):
    if not var_indices:
        features = tf.parse_single_example(serial_example,
                features=dict(zip(vars,
                                  [tf.FixedLenFeature([], tf.string)]*len(vars))))
        # 注意np.ndarray浮点数默认采用np.float64不是float32，采用float32逆编码结果错误！！！
        res = [tf.decode_raw(features[k], tf.float64) for k in vars]
    else:
        vars_ = [vars[i] for i in var_indices]
        features = tf.parse_single_example(serial_example,
                features=dict(zip(vars_,
                                  [tf.FixedLenFeature([], tf.string)] * len(var_indices))))
        # 注意np.ndarray浮点数默认采用np.float64不是float32，采用float32逆编码结果错误！！！
        res = [tf.decode_raw(features[k], tf.float64) for k in vars_]
    return res
````

直接读取**TFRecord**文件并分别创建train和test Dataset
